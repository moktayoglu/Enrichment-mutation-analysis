```python
#counting all enriching cases on ACE2
count_ace2 = 0
only_nonsyn = 2340
wt_count = 0 
for i in range(only_nonsyn):
  if (ACE2['Replicate 1'][i] + ACE2['Replicate 2'][i]) / 2 > 0:
    count_ace2 += 1
  
  if ACE2['Reads in Naive Library'][i] == 'WT':
    wt_count +=1


print(wt_count)
print(count_ace2, only_nonsyn - wt_count)
print('percent enriching of ACE: ', count_ace2/(only_nonsyn - wt_count) * 100)
```

```python
count_rbd = 0
subtract_count = 0 
for i in range(len(SARS2)):
  if SARS2['bind_avg'][i] > 0:
    count_rbd += 1
    
  if SARS2['wildtype'][i] == SARS2['mutant'][i]:
    subtract_count +=1

  if SARS2['mutant'][i] == '*':
    subtract_count +=1

print(count_rbd, len(SARS2) - subtract_count)
print('percent enriching from deep muta: ', count_rbd/(len(SARS2)-subtract_count) * 100)
```
